## data preperation

```python
dataset = dataset[['coaster_name', 
         #'Length', 'Speed', 
         'Location', 'Status', 
         #'Opening date','Type', 
         'Manufacturer',
         #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
         #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
         #'height_value', 'height_unit', 
         'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset['opening_date_clean'] = pd.to_datetime(dataset['opening_date_clean'])
```

## you can rename your column names

```python
dataset = dataset.rename(columns={'coaster_name':'Coaster_Name',
               'year_introduced':'Years_introduced',
               'latitude':'Latitude',
               'longitude':'Longitude'})
```

```python
dataset.head()
```

## checking null values 

```python
dataset.isna()
```

```python
print(dataset.shape)
dataset.isnull().sum()
```

## checking duplicate rows or not

```python
dataset.loc[dataset.duplicated(subset='Coaster_Name')]

##there are 97 duplicates oin coaster_name
```

```python
#example on how to solve duplicates
dataset.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
dataset = dataset.loc[~dataset.duplicated(subset=['Coaster_Name','Location','opening_date_clean'])]
```

```python
dataset = dataset.reset_index(drop = True)
```

```python
dataset.head()
```
